```python
import altair as alt
from vega_datasets import data

iris = data.iris.url
alt.renderers.enable('notebook')
alt.Chart(iris).mark_point().encode(
    x='petalLength:Q',
    y='petalWidth:Q',
    color='species:N',
    column='species:N'
).properties(
    width=180,
    height=180
)
```

```python
import altair as alt
from vega_datasets import data

alt.renderers.enable('notebook')
source = data.iris()

base = alt.Chart(source)
scatterplot = base.mark_point().encode(
   x='sepalLength:Q',
   y='petalWidth:Q',
   color='species:N',
   column='species:N'
)
line = base.mark_line().encode(
   x='sepalLength:Q',
   y='petalWidth:Q'
)
band = base.mark_errorband(extent='ci').encode(
    x='sepalLength:Q',
    y=alt.Y('petalWidth:Q'),
)
band + line + scatterplot
```
